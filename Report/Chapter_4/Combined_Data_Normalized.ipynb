```python
from Data_DoE_GenmAb_Ambic import *
from Data_DoE_GenmAb_CDOptiCHO import *

Glucose = pd.concat([Glucose_Amb,Glucose_CD],axis=0).reset_index(drop=True)
Glutamine = pd.concat([Glutamine_Amb,Glutamine_CD],axis=0).reset_index(drop=True)
Lactate = pd.concat([Lactate_Amb,Lactate_CD],axis=0).reset_index(drop=True)
Ammonia = pd.concat([Ammonia_Amb,Ammonia_CD],axis=0).reset_index(drop=True)
TCD = pd.concat([TCD_Amb,TCD_CD],axis=0).reset_index(drop=True)
VCD = pd.concat([VCD_Amb,VCD_CD],axis=0).reset_index(drop=True)
IgG = pd.concat([IgG_Amb,IgG_CD],axis=0).reset_index(drop=True)
```

```python
def normalizeFeatures(df):
    return df.subtract(df.mean(axis=1),axis='index').divide(df.std(axis=1),axis='index').astype(float)
```

```python
Glucose_N = normalizeFeatures(Glucose)
Glutamine_N = normalizeFeatures(Glutamine)
Lactate_N = normalizeFeatures(Lactate)
Ammonia_N = normalizeFeatures(Ammonia)
TCD_N = normalizeFeatures(TCD)
VCD_N = normalizeFeatures(VCD)
IgG_N = normalizeFeatures(IgG)
```

```python
myFeatures_N = pd.concat([Glucose_N,Glutamine_N,Lactate_N,Ammonia_N,TCD_N,VCD_N,IgG_N],axis=1).astype(float)
myOutputs_N = pd.concat([Glycan_Amb,Glycan_CD],axis=0).astype(float)/100
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.95)
X = myFeatures_N.values
X = pca.fit_transform(X)
```

```python
myFeatures_N.shape
```

```python
X.shape
```

```python
myOutputs_N.head()
```

```python
y = myOutputs_N.values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.2, 
                                                    random_state=98)
```

```python
len(X_train)
```

```python
from sklearn.neural_network import MLPRegressor
all_scores = []
max_scores = []
for i in range(1,501):
    NN_test = MLPRegressor(hidden_layer_sizes=(3,4),activation='logistic',tol = 1e-5, max_iter = 300, random_state=i,solver='lbfgs')
    NN_test.fit(X_train,y_train)
    score = NN_test.score(X_test,y_test)
    all_scores.append(score)
    if score>0.9: 
        max_scores.append(i)
        print(score,i)
    
        
```

```python
plt.plot(range(1,501),all_scores,'.')
```

```python
n, bins, patches = plt.hist(x=all_scores, bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=0.85)
```

```python
# a neural network with 3 hidden units and 4 hidden layers
NN = MLPRegressor(hidden_layer_sizes=(3,4),activation='logistic',tol = 1e-5, max_iter = 300, random_state=238,solver='lbfgs')
NN.fit(X_train,y_train)        
```

```python
NN.score(X_test,y_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
ypred = NN.predict(X_test)
mean_squared_error(y_test,ypred)
```

```python
mean_absolute_error(y_test,ypred)
```

```python
r2_score(y_test,ypred,multioutput='variance_weighted')
```

```python
from collections import defaultdict
import numpy as np
layers = [1,2,3,4,5]
units  = [3,4,5,6,7]

mydict_avgscores = defaultdict(int)

for l in layers:
    for u in units:
        avgscore = []
        for i in range(1,101):
            test_NN = MLPRegressor(hidden_layer_sizes=(u,l),activation='logistic',tol = 1e-5, max_iter = 300, random_state=i,solver='lbfgs')
            test_NN.fit(X_train,y_train)
            avgscore.append(test_NN.score(X_test,y_test))
        mydict_avgscores[(u,l)] = np.mean(avgscore)
        print(np.mean(avgscore))
```

```python
mydict_avgscores
```

```python
unit_layer = np.zeros((5,5))

for i in range(5):
    for j in range(5):
        unit_layer[i,j] = mydict_avgscores[(i+3,j+1)]
        

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
cax = ax.matshow(unit_layer)
fig.colorbar(cax)
ax.set_xticklabels(['']+['1','2','3','4','5'])
ax.set_yticklabels(['']+['3','4','5','6','7'])
plt.xlabel('Number of hidden layers',{'size':14})
plt.ylabel('Number of hidden units',{'size':14});
plt.title('Average accuracy of a Deep Neural Network using different units and layers',{'size':18})
plt.savefig('HyperParams.png');
```
