```python
NNData = GenerateTestTrain(myData.X,myData.y)
```

# Model Training

To initialize the Neural Network model, we have to feed the training data and assign the number of hidden units and the learning rate. to train it, we need to set the number of episodes and the minimum error tolerance that we need.

```python
#Train the model
myNN = NNModel.NeuralNetwork(NNData.Xtrain,NNData.ytrain,4,0.05)
myNN.train(2000,1e-4)
```

# Model Testing
To test it, we just need to feed the test data that we created.

```python
# Test the model
results = myNN.test(NNData.Xtest,NNData.ytest)
```

# Model Results

The model predicts the relative glycan distribution of a monoclonal antibody by looking at the nutrient concentration and supplements data. It returns the predicted results for each test case and also the sum of squared errors in each case. 

```python
yhat = results[1]
for i in range(len(yhat)):
    plt.figure();
    plt.bar(np.arange(1,len(yhat[0])+1),yhat[i].flatten(),0.35,label='pred');
    plt.bar(np.arange(1,len(yhat[0])+1)+0.35,NNData.ytest[i],0.35,label='actual');
    plt.xticks(np.arange(1,len(yhat[0])+1),['G0+G0F','G1+G1F','G2+G2F'])
    plt.legend();
    plt.title(f'Glycan Predictions: Mixed-{i+1}');
```
